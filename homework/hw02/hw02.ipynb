# Exercise 03

```python
signal_square = SquareSignal(freq=1100)
wave_square = signal_square.make_wave(duration=0.5, framerate=100000)
spectrum_square = wave_square.make_spectrum()
spectrum_square.plot()
wave_square.write('output/square2')
```

```python
wave_square = signal_square.make_wave(duration=0.5, framerate=10000)
spectrum_square = wave_square.make_spectrum()
spectrum_square.plot()
wave_square.write('output/square2_fold')
```

SquareSignal has component frequency : Basic Frequency、3X、5X...，but 5500HZ is larger than Nyquist frequency 5000HZ (half of sampling rate)，so 5500HZ signal was fold to 4500HZ，7700HZ was fold to 2300HZ，which is shown in figure.

The sound was diffrent as the one with high sampling rate, because folding happened when low sample rate using, cause high frequency signal loss, and many complex noise appear.

# Exercise 05

```python
signal_square = SquareSignal(freq=440)
wave_square = signal_square.make_wave(duration=0.5, framerate=10000)
spectrum_square = wave_square.make_spectrum()
spectrum_square.plot()
```

```python
signal_triangle = TriangleSignal(freq=440)
wave_triangle = signal_triangle.make_wave(duration=0.5, framerate=10000)
spectrum_triangle = wave_triangle.make_spectrum()
spectrum_triangle.plot()
```

```python
signal_sawtooth = SawtoothSignal(freq=440)
wave_sawtooth = signal_sawtooth.make_wave(duration=0.5, framerate=10000)
spectrum_sawtooth = wave_sawtooth.make_spectrum()
spectrum_sawtooth.plot()
```

```python
wave_square.write('output/square_origin.wav')
wave_triangle.write('output/triangle_origin.wav')
wave_sawtooth.write('output/sawtooth_origin.wav')
```

```python
def algorithm(spectrum):
    spectrum.hs[1:] = spectrum.hs[1:]/spectrum.fs[1:]
    spectrum.hs[0] = 0
    spectrum.scale(440) # 震幅乘上係數，拿掉這行會沒聲音 ==
```

```python
algorithm(spectrum_square)
spectrum_square.plot()
```

```python
algorithm(spectrum_triangle)
spectrum_triangle.plot()
```

```python
algorithm(spectrum_sawtooth)
spectrum_sawtooth.plot()
```

```python
spectrum_square.make_wave().write('output/square_fix.wav')
spectrum_triangle.make_wave().write('output/triangle_fix.wav')
spectrum_sawtooth.make_wave().write('output/sawtooth_fix.wav')
```

hs/fs will filter high frequency signal, which will remove aliasing signal. So sounds like only SinSignal wave.
